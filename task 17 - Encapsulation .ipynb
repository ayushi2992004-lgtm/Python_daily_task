```python
# 1 Create a class named BankAccount.
class BankAccount:
    # Constructor
    def __init__(self, owner, balance):

        # create the attributes
        self.Owner_name = str(owner)
        self.__Bank_Balance = int(balance)

    # getter : access the private attribute
    def get_balance(self):
        return self.__Bank_Balance

    # setter : update the private attribute
    def deposit(self, amount):
        if amount > 0:
            self.__Bank_Balance += amount
            return self.__Bank_Balance
        else:
            return "Please Enter Valid Amount"

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__Bank_Balance:
            self.__Bank_Balance -= amount
            return self.__Bank_Balance
        else:
            return "Insufficient Balance"

    # method 1 - Instance Method
    def display_bank_details(self):
        print(f"Owner Name : {self.Owner_name}")
        print(f"Bank Balance : {self.__Bank_Balance}")


# create the class object
bank1 = BankAccount("Ayushi", 2000)

# call the methods
bank1.display_bank_details()

print("After Deposit :", bank1.deposit(1000))
print("After Withdraw :", bank1.withdraw(500))
print("Current Balance :", bank1.get_balance())
```

```python
# 16 Create an Abstract method
from abc import ABC, abstractmethod

# Abstract or Parent Class
class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass

    @abstractmethod
    def calculate_perimeter(self):
        pass


# Child Class
class Circle(Shape):

    # Constructor
    def __init__(self, radius):

        # create the attribute
        self.Radius = radius

    # method 1
    def calculate_area(self):
        return 3.14 * self.Radius * self.Radius

    # method 2
    def calculate_perimeter(self):
        return 2 * 3.14 * self.Radius


# Child Class
class Rectangle(Shape):

    # Constructor
    def __init__(self, length, width):

        # create the attributes
        self.Length = length
        self.Width = width

    # method 1
    def calculate_area(self):
        return self.Length * self.Width

    # method 2
    def calculate_perimeter(self):
        return 2 * (self.Length + self.Width)


# create the class objects
circle1 = Circle(7)
rectangle1 = Rectangle(10, 5)

# call the methods
print("Circle Area :", circle1.calculate_area())
print("Circle Perimeter :", circle1.calculate_perimeter())

print("Rectangle Area :", rectangle1.calculate_area())
print("Rectangle Perimeter :", rectangle1.calculate_perimeter())
```

```python
# 25 Create abstract class Shape3D.

# Abstract Class
class Shape3D(ABC):

    @abstractmethod
    def volume(self):
        pass

    @abstractmethod
    def surface_area(self):
        pass

# Child Class
class Cube(Shape3D):

    # Constructor
    def __init__(self, side):

        # create the attribute
        self.Side = side

    # method 1
    def volume(self):
        return self.Side ** 3

    # method 2
    def surface_area(self):
        return 6 * (self.Side ** 2)

# Child Class
class Sphere(Shape3D):

    # Constructor
    def __init__(self, radius):

        # create the attribute
        self.Radius = radius

    # method 1
    def volume(self):
        return (4/3) * 3.14 * (self.Radius ** 3)

    # method 2
    def surface_area(self):
        return 4 * 3.14 * (self.Radius ** 2)

# Child Class
class Cylinder(Shape3D):

    # Constructor
    def __init__(self, radius, height):

        # create the attributes
        self.Radius = radius
        self.Height = height

    # method 1
    def volume(self):
        return 3.14 * (self.Radius ** 2) * self.Height

    # method 2
    def surface_area(self):
        return 2 * 3.14 * self.Radius * (self.Radius + self.Height)

# create the class objects
cube1 = Cube(5)
sphere1 = Sphere(7)
cylinder1 = Cylinder(3, 8)

# call the methods
print("Cube Volume :", cube1.volume())
print("Cube Surface Area :", cube1.surface_area())

print("Sphere Volume :", sphere1.volume())
print("Sphere Surface Area :", sphere1.surface_area())

print("Cylinder Volume :", cylinder1.volume())
print("Cylinder Surface Area :", cylinder1.surface_area())
```
