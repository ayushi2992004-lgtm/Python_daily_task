```python
# 1 Create an abstract class named Shape and subclasses like circle and Ractangle and implement both method.
from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


# Child Class
class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius


# Child Class
class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


# Objects
circle = Circle(5)
rectangle = Rectangle(10, 4)

print("Circle")
print("Area =", circle.area())
print("Perimeter =", circle.perimeter())

print()

print("Rectangle")
print("Area =", rectangle.area())
print("Perimeter =", rectangle.perimeter())
```

```python
# 31 Create an abstract class named BankAccount and subclass as Saving account, CurrentAccount. A account store the account balance, but different account types calculate interest differently.

# Abstract Class
class BankAccount(ABC):

    # Class Variable
    total_accounts = 0

    # Constructor
    def __init__(self, balance):
        self.balance = balance
        BankAccount.total_accounts += 1

    # Abstract Method
    @abstractmethod
    def calculate_interest(self):
        pass

    # Class Method
    @classmethod
    def show_total_accounts(cls):
        print("Total Bank Accounts :", cls.total_accounts)


# Child Class
class SavingsAccount(BankAccount):

    MIN_BALANCE = 1000

    def __init__(self, balance):
        super().__init__(balance)

    def calculate_interest(self):
        interest = self.balance * 5 / 100
        print("Savings Interest :", interest)

    def withdraw(self, amount):
        if self.balance - amount >= SavingsAccount.MIN_BALANCE:
            self.balance -= amount
            print("Withdraw Successful")
            print("Remaining Balance :", self.balance)
        else:
            print("Minimum Balance should be ₹1000")


# Child Class
class CurrentAccount(BankAccount):

    OVERDRAFT_LIMIT = -5000

    def __init__(self, balance):
        super().__init__(balance)

    def calculate_interest(self):
        print("Current Account gives No Interest")

    def withdraw(self, amount):
        if self.balance - amount >= CurrentAccount.OVERDRAFT_LIMIT:
            self.balance -= amount
            print("Withdraw Successful")
            print("Current Balance :", self.balance)
        else:
            print("Overdraft Limit Exceeded")

# Objects
acc1 = SavingsAccount(10000)
acc2 = SavingsAccount(2500)
acc3 = CurrentAccount(5000)
acc4 = CurrentAccount(1000)

print("Savings Account")
acc1.calculate_interest()
acc1.withdraw(8500)

print()

print("Current Account")
acc3.calculate_interest()
acc3.withdraw(8000)

print()

print("Current Account")
acc4.withdraw(7000)

print()

BankAccount.show_total_accounts()
```

```python
# 32 Create an abstract class named Shape and subclasses Rectangle, Circle. Every shape has a color, but different shapes calculate area and perimeter differently. Create a constructor:

# Abstract Class
class Shape(ABC):

    def __init__(self, color):
        self.color = color

    # Normal Method
    def display_color(self):
        print("Color :", self.color)

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


# Child Class
class Rectangle(Shape):

    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def display_details(self):
        super().display_color()
        print("Length :", self.length)
        print("Width :", self.width)
        print("Area :", self.area())
        print("Perimeter :", self.perimeter())


# Child Class
class Circle(Shape):

    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius

    def display_details(self):
        super().display_color()
        print("Radius :", self.radius)
        print("Area :", self.area())
        print("Perimeter :", self.perimeter())


# Objects
r = Rectangle("Blue", 10, 5)
c = Circle("Red", 7)

r.display_details()

print()

c.display_details()
```

```python
# 34 Create an abstract class named Vehicle and subclasses ElectricCar, PetrolCar Every vehicle has common information like brand and fuel type, but different vehicles start and consume fuel differently.

class Vehicle(ABC):

    def __init__(self, brand, fuel_type):
        self.brand = brand
        self.fuel_type = fuel_type

    def display_vehicle(self):
        print("Brand :", self.brand)
        print("Fuel Type :", self.fuel_type)

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def fuel_consumption(self, distance):
        pass


class ElectricCar(Vehicle):

    def __init__(self, brand, fuel_type):
        if fuel_type != "Electric":
            raise ValueError("Fuel type must be Electric")
        super().__init__(brand, fuel_type)

    def start(self):
        print("Electric Car Started Silently")

    def fuel_consumption(self, distance):
        print("Battery Used :", distance * 0.15, "kWh")

    def display_vehicle(self):
        super().display_vehicle()
        print("Vehicle Type : Electric Car")


class PetrolCar(Vehicle):

    def __init__(self, brand, fuel_type):
        if fuel_type != "Petrol":
            raise ValueError("Fuel type must be Petrol")
        super().__init__(brand, fuel_type)

    def start(self):
        print("Petrol Car Started")

    def fuel_consumption(self, distance):
        print("Petrol Used :", distance / 15, "Litres")

    def display_vehicle(self):
        super().display_vehicle()
        print("Vehicle Type : Petrol Car")


# Objects
e = ElectricCar("Tesla", "Electric")
p = PetrolCar("Toyota", "Petrol")

e.display_vehicle()
e.start()
e.fuel_consumption(100)

print()

p.display_vehicle()
p.start()
p.fuel_consumption(150)
```

```python
# 35 Create an abstract class named Report and subclasses as PDFReport, HTMLReport. Every report has common information like title and data, but different report formats generate output differently.

class Report(ABC):

    def __init__(self, title, data):
        self.title = title
        self.data = data

    def header(self):
        print("===== REPORT =====")
        print("Title :", self.title)

    def footer(self):
        print("===== END OF REPORT =====")

    @abstractmethod
    def generate(self):
        pass


class PDFReport(Report):

    def __init__(self, title, data):
        super().__init__(title, data)

    def generate(self):
        super().header()
        print("Generating PDF Report...")
        print("Data :", self.data)
        super().footer()


class HTMLReport(Report):

    def __init__(self, title, data):
        super().__init__(title, data)

    def generate(self):
        super().header()
        print("Generating HTML Report...")
        print("Data :", self.data)
        super().footer()


# Objects
pdf = PDFReport("Student Report", "Marks Data")
html = HTMLReport("Employee Report", "Salary Data")

pdf.generate()

print()

html.generate()
```
