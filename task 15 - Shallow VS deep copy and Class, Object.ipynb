```python
# 1 Create a list original = [1, 2, [3, 4], 5]. Make a shallow copy named shallow and change shallow[2][0] = 99. Print both lists and explain the result.
import copy
original = [1, 2, [3, 4], 5]
shallow = copy.copy(original)
shallow[2][0] = 99
print("Original List: ", original)
print("Shallow Copy: ", shallow)
```

```python
# 2 Do the same as above but use deepcopy instead of shallow copy. Show the difference.
import copy
original = [1, 2, [3, 4], 5]
deep = copy.deepcopy(original)
deep[2][0] = 99
print("Original List: ", original)
print("Deep Copy: ", deep)
```

```python
# 3 Write a program to demonstrate shallow copy using copy.copy() on a nested list.
import copy
nested_list = [[10, 20], [30, 40], [50, 60]]
shallow = copy.copy(nested_list)
print("Before Modification")
print("Original: ", nested_list)
print("Shallow: ", shallow)
shallow[1][0] = 99
print("\nAfter Modification")
print("Original: ", nested_list)
print("Shallow: ", shallow)
```

```python
# 4 Write a program to demonstrate deep copy using copy.deepcopy() on the same nested list.
import copy
nested_list = [[10, 20],[30, 40], [50, 60]]
deep = copy.deepcopy(nested_list)
print("Before Modification")
print("Original: ", nested_list)
print("Deep Copy: ", deep)
deep[1][0] = 99
print("\nAfter Modification")
print("Original: ", nested_list)
print("Deep Copy: ", deep)
```

```python
# 7 Write a function manual_deep_copy(lst) that manually creates a deep copy of a nested list without using copy.deepcopy().
def manual_deep_copy(lst):

    if isinstance(lst,list):
        return [manual_deep_copy(item) for item in lst]

    else:
        return lst

original = [1, 2, [3, 4], [5, [6, 7]]]
copy_list = manual_deep_copy(original)

# update the orginal list 

original[3][1][1]  = 100

print("Orginal List :",original)
print("Copy List : ",copy_list)
```

```python
# 9 Create a function that detects whether a copy is shallow or deep by modifying a nested element and checking the original.
import copy
def detect_copy(Original, Copied):
    Copied[1][0] = 999
    if Original[1][0] == 999:
        print("This is a Shallow Copy")
    else:
        print("This is a Deep Copy")
data = [10, [20, 30], 40]
shallow = copy.copy(data)
deep = copy.deepcopy(data)
print("Checking Shallow Copy: ")
detect_copy(data, shallow)

# Reset Original data
data = [10, [20, 30], 40]
print("\nChecking Deep Copy: ")
deep = copy.deepcopy(data) 
detect_copy(data, deep)
```

```python
# 15 Bank Account
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
    def deposite(self, amount):
        self.balance += amount
        print("Deposite Successfully", amount)
    def withdraw(self,amount):
        if amount>0:
            if amount <= self.balance:
                self.balance -= amount
                print("Withdraw Successfully!", amount)
            else:
                print("Insufficient Balance!")
        else:
            print("Please Enter the Amount Greater than Zero")
    def display(self):
        print("Account Holder: ", self.account_holder)
        print("Balance: ",self.balance)

account = BankAccount("Rahul",20000)
print("Initial Account Details")
account.display()
print()
account.deposite(5000)
print()
account.withdraw(10000)
print()
account.display()
```

```python
# 35 Online Shopping Cart
class OnlineShoppingCart:
    def __init__(self, customer_name, wallet_balance):
        self.Customer_Name = customer_name
        self.Cart_Total = 0
        self.Wallet_Balance = wallet_balance
    def add_item(self, price):
        self.Cart_Total += price
        print("Item Added Successfully")
        print("Cart Total =", self.Cart_Total)
    def remove_item(self, price):

        if price > self.Cart_Total:
            print("Cart Total Cannot Become Negative")
        else:
            self.Cart_Total -= price
            print("Item Removed Successfully")
            print("Cart Total =", self.Cart_Total)

    def checkout(self):

        if self.Cart_Total == 0:
            print("Cart is Empty")

        elif self.Cart_Total <= self.Wallet_Balance:
            self.Wallet_Balance -= self.Cart_Total
            print("Payment Successful")
            print("Wallet Balance =", self.Wallet_Balance)
            self.Cart_Total = 0
            print("Cart Total =", self.Cart_Total)

        else:
            print("Insufficient Wallet Balance")

cart = OnlineShoppingCart("Ayushi", 5000)

cart.add_item(1000)
cart.add_item(1500)

cart.remove_item(500)

cart.checkout()
```

```python
# 36 Create a University Student Management System.
class Student:
    def __init__(self, student_id, name, age, course, total_fee, paid_fee,
                 marks, attended_classes, total_classes):

        self.Student_ID = student_id
        self.Student_Name = name
        self.Age = age
        self.Course = course
        self.Total_Fee = total_fee
        self.Paid_Fee = paid_fee
        self.Marks = marks
        self.Attended_Classes = attended_classes
        self.Total_Classes = total_classes

    def update_marks(self, marks):

        if 0 <= marks <= 100:
            self.Marks = marks
            print("Marks Updated Successfully")
        else:
            print("Invalid Marks")

    def pay_fee(self, amount):

        if self.Paid_Fee + amount <= self.Total_Fee:
            self.Paid_Fee += amount
            print("Fee Paid Successfully")
        else:
            print("Fee exceeds Total Fee")

    def update_attendance(self, classes):

        if self.Attended_Classes + classes <= self.Total_Classes:
            self.Attended_Classes += classes
            print("Attendance Updated")
        else:
            print("Attendance Cannot Exceed Total Classes")

    def change_course(self, new_course):

        self.Course = new_course
        print("Course Changed Successfully")

    def attendance_percentage(self):

        percentage = (self.Attended_Classes / self.Total_Classes) * 100
        return percentage

    def check_result(self):

        if self.Marks >= 35:
            return "Pass"
        else:
            return "Fail"

    def scholarship(self):

        if (self.Marks > 90 and
            self.attendance_percentage() > 95 and
            self.Paid_Fee == self.Total_Fee):

            return "Yes"

        else:
            return "No"

    def display_student_report(self):

        print("\n========== Student Report Card ==========")
        print("Student ID :", self.Student_ID)
        print("Name       :", self.Student_Name)
        print("Age        :", self.Age)
        print("Course     :", self.Course)
        print("Marks      :", self.Marks)
        print("Result     :", self.check_result())

        print("Attendance :", self.attendance_percentage(), "%")

        print("Fee Status :", self.Paid_Fee, "/", self.Total_Fee)

        print("Scholarship Eligible :", self.scholarship())

s1 = Student(101,"Ayushi",18,"B.Tech AI",50000,30000,88,90,100)

s1.update_marks(95)
s1.pay_fee(20000)
s1.update_attendance(6)
s1.change_course("B.Tech Data Science")

s1.display_student_report()
```
