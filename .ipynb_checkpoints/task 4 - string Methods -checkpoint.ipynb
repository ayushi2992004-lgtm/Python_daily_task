```python
# 5 Split comma-separated string
fruits = input("Enter fruits seperated by commas: ")
fruits.split(",")
print(fruits)
```

```python
# 6 Join list of Fruits with space
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]
print(" ".join(fruits))
```

```python
# 7 Strip each word and join
words = ["Python", "is", "awesome"]
clean = [word.strip() for word in words]
print(" ".join(clean))
```

```python
# 11 Replace and split
text1 = "Hello world python"
text2 = text1.replace("World","Everyone")
print(text2.split())
```

```python
# 12 Join city names with new line
cities = []
for i in range(5):
    city = input("Enter city: ")
    cities.append(city)
print("\n".join(cities))
```

```python
# 13 Strip, title case, and check startswith
text = "Hello world"
text = text.strip().title()
print(text)
print(text.startswith("Hello"))
```

```python
# 14 Replace "bad" with "Good" after spliting and joining
sentence = input("Enter a sentence: ")
words = sentence.split()
sentence = " ".join(words)
print(sentence.replace("bad","good"))
```

```python
# 15 Join list with "_"
number = ["10","20","30","40"]
print("_".join(number))
```

```python
# 16 Remove words starting with vowels
sentence = input("Enter sentence: ")
words = sentence.split()
result = []
for word in words:
    if word[0].lower() not in "aeiou":
        result.append(word)
print(" ".join(result))
```

```python
# 17 Replace Space with"_", split and print words>4 Letters
text = input("Enter string: ")
text = text.replace(" ", "-")
words = text.split("-")
for word in words:
    if len(word) > 4:
        print(word)
```

```python
# 18 Replace First Student Name
students = ["Ram", "Amit", "Neha", "Raj"]
print(", ". join(students))
students[0] = "New Student"
print(", ".join(students))
```

```python
# 19 split paragraph into sentences
paragraph = input("Enter Paragraph: ")
sentences = paragraph.split(".")
result = []
for sentence in sentences:
    sentence = sentence.strip()
    if sentence != "":
        result.append(sentence)
print("- "+ "\n- ".join(result))
```

```python
# 20 Clean and Reformat sentence
text = input("Enter sentence: ")
text = text.strip()
text = text.replace(",", "")
words = text.split()
result = " ".join(words)
print(result)
```
