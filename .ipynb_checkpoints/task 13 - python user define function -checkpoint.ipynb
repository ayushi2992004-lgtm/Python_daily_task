```python
# 31 function process_data(*args, **kwargs)
def process_data(*args,** kwargs):
    result = {}
    if kwargs.get("add"):
        result["Addition"] = sum(args)
    if kwargs.get("multiply"):
        product = 1
        for i in args:
            product *= i
            result["Multiplication"] = product
    if kwargs.get("power"):
        result["Power"] = [i**2 for i in args]
    return result
print(process_data(2, 3, 4, add=True, multiply = True, power = True))
```

```python
# 32 create double = make_multiplier(2) and triple = make_multiplier(3)
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply
double = make_multiplier(2)
triple = make_multiplier(3)
print(double(5))
print(triple(5))
```

```python
# 33 function smart_divide(func) as a decorator that:
def smart_divide(func):
    def wrapper(a,b):
        if b == 0:
            print("Cannot divide by Zero!")
            return
        return func(a,b)
    return wrapper
@smart_divide
def divide(a,b):
    return a/b
print(divide(10,2))
print(divide(10,0))
```

```python
# 34 function flexible_calc(*args, operation="add", **kwargs)
def flexible_calc(*args, operation="add", round_result=False):

    if operation == "add":
        result = sum(args)

    elif operation == "subtract":
        result = args[0]
        for i in args[1:]:
            result -= i

    elif operation == "multiply":
        result = 1
        for i in args:
            result *= i

    elif operation == "divide":
        result = args[0]
        for i in args[1:]:
            if i == 0:
                return "Division by zero not allowed"
            result /= i

    else:
        return "Invalid Operation"

    if round_result:
        return round(result, 2)

    return result


print(flexible_calc(10, 5, 2, operation="add"))
print(flexible_calc(10, 5, 2, operation="subtract"))
print(flexible_calc(10, 5, 2, operation="multiply"))
print(flexible_calc(10, 5, 2, operation="divide"))
```

```python
# 35 recursive function flatten_list(nested_list) that takes a nested list (list inside list) and returns a single flat list.
def flatten_list(nested_list):
    result = []
    for item in nested_list:
        if isinstance(item, list):
            result.extend(flatten_list(item))
        else:
            result.append(item)
    return result
data = [1, [2,3],[4, [5, 6]]]
print(flatten_list(data))
```
