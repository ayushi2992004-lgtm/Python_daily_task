```python
# 22 Flatten this nested list using list comprehension: [[1,2],[3,4],[5,6]]
Nested = [[1,2], [3,4], [5,6]]
flat = [num for sublist in Nested for num in sublist]
print(flat)
```

```python
# 29 list comprehension that combines two lists with a condition.
list1 = [1,2,3,4]
list2 = ['A', 'B', 'C', 'D']
result = [(x,y) for x in list1 for y in list2 if x%2==1]
print(result)
```

```python
# 30 difference in readability and performance between a normal for loop and list comprehension.
# square using for loop
square = []
for i in range(1,6):
    square.append(i**2)
print(square)
```

```python
# 1 Combined Functions + List Comprehension Write a single line of code using list comprehension + built-in functions to:
result = [str(i)+ "(multiple)" for i in range(1, 51) if i % 3 == 0 and i % 5 == 0]
print(result)
```

```python
# 2 Hard Type Conversion + Logical Functions Given this messy list:
data = ["10", "abc", "25.5", "30", "45.0", "xyz", "60"]
result = []
for item in data:
    try:
        result.append(int(float(item)))
    except:
        pass
print(result)    
    
```

```python
# 3 Advanced Enumeration + Zip + Condition You have two lists:
name = ["Sumit", "Rahul", "Priya", "Aman", "Neha"]
score = [85,92,78,95,88]
result = [(index,name,"Excellent" if score>=90 else "Good") for index, (name,score) in enumerate(zip(name, score))]
print(result)
```

```python
# 4 Nested + Multiple Built-ins Write a list comprehension to generate a multiplication table (5x5)
table = [
    [
        str(i*j) + ("(Ok)" if (i*j)%2 == 0 else "(Odd)")
        for j in range (1, 6)
        if(i*j) % 7 != 0
    ]
    for i in range(1, 6)
]
print(table)
```

```python
# 5 Memory + Logic + Conversion Predict and explain the output of this code, then optimize it into a clean single-line list comprehension:
result = []
for i in range(1,21):
    if i % 2 == 0:
        result.append(str(pow(i, 2)) + " → " + str(abs(i - 10)))
print(result)
```
