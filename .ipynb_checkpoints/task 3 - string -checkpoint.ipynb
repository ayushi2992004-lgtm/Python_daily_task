```python
# String Slicing Questions
# 1. Take a string from user and print the first 5 characters.
text = input("Enter a Sentence: ")
print(text[:5])
```

```python
# 2. Take a string and print the last 4 characters using negative slicing.
text = input("Enter a Sentence: ")
print(text[-4:])
```

```python
# 3. Take a string and print characters from index 2 to 8 (excluding 8).
text = input("Enter a Sentence: ")
print(text[2:8])
```

```python
# 4. Take a string and print every 2nd character of the string.
text = input("Enter a Sentence: ")
print(text[::2])
```

```python
# 5. Take a full name (e.g. "Sumit Kumar Sharma") and print only the first name and last name separately using slicing.
name = input("Enter Full Name: ")
space = name.find(" ")
print("First Name: ", name[:space])
print("Second Name: ", name[space+1:])
```

```python
# Case Conversion Questions
# 6. Take a string from user and print it
text = input("Enter a Sentence: ")
print("Uppercase: ", text.upper())
print("Lowercase: ", text.lower())
print("Title case: ", text.title())
print("Capitization: ", text.capitalize())
```

```python
# 7. Take a sentence and convert it to title case. Then print the original and converted version.
sentence = input("Enter a Sentence: ")
print("Original: ", sentence)
print("Title: ", sentence.title())
```

```python
# 8. Take a mixed case string and check if lower() and upper() methods change the string permanently or not.
word = input("Enter Character: ")
print(word.upper())
print(word.lower())
print("Original: ", word)
```

```python
# String Check Methods
# 9. Take a string from user and check whether it contains only alphabets (isalpha()).
ch = input("Enter Character: ")
print(ch.isalpha())
```

```python
# 10. Take input from user and check if it is a digit only (isdigit()).
text = input("Enter Input: ")
print(text.isdigit())
```

```python
# 11. Take a string and check if it is alphanumeric (isalnum()).
text = input("Enter Input: ")
print(text.isalnum())
```

```python
# 12. Take a string and check if it is completely in lowercase (islower()).
text = input("Enter Character: ")
print(text.islower())
```

```python
# 13. Take a password as input and check:
password = input("Enter Password: ")
if password.isalnum() and len(password)>=8:
    print("Valid Password")
else:
    print("Invalid Password")
```

```python
# find() and index()
# 14. Take a string and a character from user. Find the first occurrence of that character using find().
text = input("Enter String: ")
ch = input("Enter Character: ")
print(text.find(ch))
```

```python
# 15. Take a string and a word. Use find() to check if the word exists. If not found, print -1.
text = input("Enter Sentence: ")
word = input("Enter word: ")
print(text.find(word))
```

```python
# 16. Take a sentence and find the position of the word "Python" using index().
sentence = input("Enter sentence: ")
print(sentence.index("Python"))
```

```python
# 17. Take a string. Use both find() and index() to search for a character that does not exist. Observe the difference.
text = input("Enter Text: ")
print(text.find("z"))
print(text.index("z"))
```

```python
# 18. Take your full name.
name = input("Enter Full Name: ")
print(name.upper())
space = name.find(" ")
print("Space Position: ", space)
print("First Name: ", name[:space])
print("Last Name: ", name[space+1:])
```

```python
# 19. Take a mobile number as string. Check if it is 10 digits and contains only numbers using isdigit() and len(). 
mobile = input("Enter Mobile Number: ")
if len(mobile) == 10 and mobile.isdigit():
    print("Valid Mobile Number")
else:
    print("Invalid Mobile Number")
```

```python
# 20. Reverse, Title Case and palindrome Check
text = input("Enter String: ")
print("Reverse: ", text[::-1])
print("Title: ", text.title())
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
