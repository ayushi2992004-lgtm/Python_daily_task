```python
# 19 Given data = [10, 20, 30, 20, 40, 20], remove all occurrences of 20 (use loop or multiple remove).
list =[10, 20, 30, 20, 40, 20]
while 20 in list:
    list.remove(20)
print(list)
```

```python
# 20 Create a list of student marks. Sort it, find the highest and lowest mark using list methods, and print them.
marks = [98, 72, 56, 88.4, 65, 52]
marks.sort()
print("Sorted marks: ", marks)
print("Lowest marks: ", marks[0])
print("Highest marks: ", marks[-1])
```
