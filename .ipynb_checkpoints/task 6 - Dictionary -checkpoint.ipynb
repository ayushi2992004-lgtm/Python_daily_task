```python
#22 Tuple operation
tup = (1, 5, 10, 15, 20, 25, 30, 25)
print(tup.index(25))
print(tup.count(25))
list2 = list(tup)
print(list2)
list2.sort(reverse=True)
print(list2)
```

```python
# 23 store 4 Students and print highest marks
student = {}
for i in range(4):
    Name=("Enter the name of Student :")
    Marks=("Enter the marks of Students :")
    student[Name]=Marks
highest = max(student, key=student.get)
print("Top Student: ",highest)
print("Marks: ",student[highest])
```

```python
# 24 copy, clear original and add country
info = {"name":"Aman","age":21,"city":"Mumbai"}
copy_info = info.copy()
info.clear()
copy_info["Country"]="India"
print("Original: ",info)
print("Copy: ",copy_info)
```

```python
# 25 count occurrences using tuple and dictionary
data = (10, 20, 30, 20, 40, 20, 50)
result = {}
for num in data:
    result[num]=data.count(num)
print(result)
```
