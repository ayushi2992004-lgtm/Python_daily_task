```python
# 1 Take a number as input. If the number is positive, check whether it is even or odd and print the result.
num = int(input("Enter a number: "))
if num>0:
    if num%2==0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Not Positive")
```

```python
# 2 Take age as input. If age is greater than or equal to 18, check if the person is a senior citizen (age >= 60) or adult.
age = int(input("Enter age: "))
if age>=18:
    if age>=60:
        print("Senior Citizen")
    else:
        print("Adult")
else:
    print("Minor")
```

```python
# 3 Take a single character. If it is an alphabet, check whether it is uppercase or lowercase.
Ch = input("Enter a Character: ")
if Ch.isalpha():
    if Ch.isupper():
        print("Uppercase")
    else:
        print("Lowercase")
else:
    print("Not an Alphabet")
```

```python
# 4 Take three numbers as input and print the largest number using nested if statements (without using max()).
a = int(input("Enter First Number"))
b = int(input("Enter Second Number"))
c = int(input("Enter Third Number"))
if a >= b:
    if a >= c:
        print("Greater number: ",a)
    else:
        print("Greater number: ",c)
else:
    if b >= c:
        print("Greater number: ",b)
    else:
        print("Greater number: ",c)
```

```python
# 8 Take three sides of a triangle. First check if they form a valid triangle, then check if it is Equilateral, Isosceles, or Scalene.
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("Equilateral triangle")
    elif a==b or b==c or a==c:
        print("Isosceles")
    else:
        print("Scalene")
else:
    print("Invalid Traingle")
```

```python
# 10 Take a character. If it is a letter, check if it is vowel or consonant. If it is a digit, print "Digit". Else print "Special Character".
text = input("Enter Character: ")
if text.isalpha():
    if text.lower() in ['a', 'e', 'i', 'o', 'u']:
        print("Vowel")
    else:
        print("Consonant")
elif text.isdigit():
    print("Digit")
else:
    print("Special Character")
```

```python
# 11 Take a year as input. Check if it is a leap year using nested conditions.
year = int(input("Enter year: "))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print("Leap Year")
        else:
            print("Not Leap Year")
    else:
        print("Leap Year")
else:
     print("Not Leap Year")   
```

```python
# 18 Take three sides of a triangle. Check if valid, then check if it is a right-angled triangle using Pythagoras theorem.
a = int(input("Enter length of side = "))
b = int(input("Enter length of side = "))
c = int(input("Enter length of side = "))
if a+b>c and a+c>b and b+c>a:
    sides = sorted([a,b,c])
    if sides[0]**2 + sides[1]**2 == sides[2]**2 :
        print("Right Angle Triangle")
    else:
        print("Not a Right Angle Triangle")
else:
    print("Not a Valid Trianle")
```

```python
# 21 Types of Triangle
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("Equilateral triangle")
    elif a==b or b==c or a==c:
        print("Isosceles triangle")
    else:
        print("Scalene triangle")
else:
    print("Triangle is not Valid")
        
        
```
