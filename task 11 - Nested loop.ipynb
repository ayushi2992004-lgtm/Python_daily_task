```python
# 11 Reverse a number using while loop.
num = int(input("Enter a Number: "))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
print("Reverse Number: ",reverse)
```

```python
# 12 Check if a number is prime using while loop.
num = int(input("Enter a Number: "))
i = 2
prime = True
while i < num:
        if num % i == 0:
            prime = False
            break
        i += 1
if prime and num > 1:
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
# 13 Print Fibonacci series up to n terms using while loop.
a = 0
b = 1
count = 1
while count <= 8:
    print(a)
    c = a+b
    a = b
    b = c
    count += 1 
```

```python
# 22 Print a hollow square pattern of size n using nested loop.
n = int(input("Enter a Number: "))
for i in range (n):
    for j in range (n):
        if i == 0 or i == n-1 or j == 0 or j == n - 1:
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
# 23 Print Floyd’s Triangle using nested for loop.
n = int(input("Enter the Number of Rows: "))
num = 1
for i in range (1, n+1):
    for j in range (i):
        print(num, end = " ")
        num += 1
    print()
```

```python
# 34 Simulate a login system: 3 attempts, use break on success.
username = "admin"
password = "1234"
attempt = 1
while attempt <= 3:
    user = input("Enter Username: ")
    pwd = input("Enter Password: ")
    if user == username and pwd == password:
        print("Login Successfully")
        break
    else:
        print("Wrong Username or Password")
        attempt += 1
if attempt >= 3:
    print("Login Failed")
```
