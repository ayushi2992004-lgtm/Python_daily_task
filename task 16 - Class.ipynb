```python
# 13 In class Circle, add class attribute pi = 3.14. Use it in area calculation method.
class Circle:
    pi = 3.14
    def __init__(self, radius):
        self.Radius = radius
    def calculate_area(self):
        area = Circle.pi * self.Radius * self.Radius
        print(f"Radius = {self.Radius}")
        print(f"Area = {area}")
c1 = Circle(7)
c1.calculate_area()
```

```python
# 26 Define class Temperature with private __celsius. Add getter and setter with conversion.
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius
    def get_celsius(self):
        return self.__celsius
    def set_celsius(self,celsius):
        self.__celsius = celsius
        print("Temperature Updated Successfully")
    def convert_to_fahrenheit(self):
        fahrenheit = (self.__celsius*9/5)+32
        print(f"Farenheit = {fahrenheit}")
t1 = Temperature(25)
print("Celsius= ",t1.get_celsius())
t1.convert_to_fahrenheit()
t1.set_celsius(40)
print("Updated Celsius = ",t1.get_celsius())
t1.convert_to_fahrenheit()
```

```python
# 28 In class Product, add getter and setter for private __price with discount logic.
class Product:
    def __init__(self, price):
        self.__price = price
    def get_price(self):
        return self.__price
    def set_price(self, price):
        if price > 5000:
            price = price - (price * 10 / 100)
            print("10% Discount Applied")
        self.__price = price
p1 = Product(4000)
print("Old Price =", p1.get_price())
p1.set_price(6000)
print("New Price =", p1.get_price())
```

```python
# 30 Define class Rectangle with private __length and __width. Add getter/setter and area method.
class Rectangle:
    def __init__(self, length, width):
        self.__length = length
        self.__width = width
    def get_length(self):
        return self.__length
        
    def get_width(self):
        return self.__width

    def set_length(self, length):
        self.__length = length
        print("Length Updated Successfully")

    def set_width(self, width):
        self.__width = width
        print("Width Updated Successfully")

    def calculate_area(self):
        area = self.__length * self.__width
        print("Area =", area)

r1 = Rectangle(10, 5)

print("Length =", r1.get_length())
print("Width =", r1.get_width())

r1.set_length(20)
r1.set_width(8)

print("Updated Length =", r1.get_length())
print("Updated Width =", r1.get_width())

r1.calculate_area()
```

```python
# 31 Create class Employee with private attribute __salary. Add getter and setter with these conditions in setter:
class Employee:
    total_employees = 0
    last_updated_salary = 0
    def __init__(self, name, salary):
        self.Employee_Name = name
        self.__salary = salary
        Employee.total_employees += 1

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):

        if salary < 5000:
            print("Salary cannot be less than 5000")
            return

        if salary > self.__salary * 1.20:
            print("Salary cannot be increased by more than 20%")
            return

        self.__salary = salary
        Employee.last_updated_salary = salary
        print("Salary Updated Successfully")

    def display_employee_details(self):
        print(f"Employee Name = {self.Employee_Name}")
        print(f"Salary = {self.__salary}")

    @classmethod
    def get_total_employees(cls):
        print("Total Employees =", cls.total_employees)

    # Class Method
    @classmethod
    def get_last_updated_salary(cls):
        print("Last Updated Salary =", cls.last_updated_salary)

emp1 = Employee("Ayushi", 10000)
emp2 = Employee("Rahul", 15000)

emp1.display_employee_details()
print()

emp1.set_salary(12000)
print()

emp1.set_salary(16000)
print()

emp1.set_salary(4000)
print()

emp1.display_employee_details()
print()

Employee.get_total_employees()
Employee.get_last_updated_salary()
```
