```python
# 1 Addmission Eligiblity
age=int(input("Enter the age of Student :"))
marks=int(input("Enter the marks Obtained: "))

if (18 < age < 25 ) and marks >= 60:
    print("eligible")
else:
    print("not eligible")
```

```python
# 2 Login System
username=input("Enter the Username :")
password=input("Enter the Password : ")
if username == "admin" and password == "12345":
    print("Login Succesful")
else:
    print("Login Failed")
```

```python
# 3 Number Between 50 and 100
num=int(input("Enter the number :"))
if num>=50 and num<=100 and not (num % 5 == 0):
    print("Valid Number")
else:
    print("Invalid Number")
```

```python
# 4 largest of three numbers
a=input("Enter first number :")
b=input("Enter second number :")
c=input("Enter third number :")
if a >= b and a >= c:
    print("Largest = ", a)
elif b >=  a and b >= c:
    print("Largest =", b)
else:
    print("Lagrest = ", c)
```

```python
# 5 Full Name Operation
name = input("Enter your full name: ")
print("Uppercase: ", name.upper())
print("Lowercase: ", name.lower())
print("Length: ",len(name))
print("First 3 Characters: ", name[:3])
print("Last 3 Character: ", name[-3:])
```

```python
# 6 Sentence Slicing
sentence = input("Enter a sentence: ")
print("First 10: ",sentence[:10])
print("Last 10: ", sentence[-10:])
print("Index 5 to 15: ", sentence[5:16])
print("Every 2nd character: ", sentence[::2])
```

```python
# 7 Palindrome
word = input("Enter a word: ")
if word == word[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 8 Email Username and Domain
email = input("Enter email: ")
parts = email.split("@")
print("Username: ",parts[0])
print("Domain: ",parts[1])
```

```python
# 9 strong password
password = input("Enter Password: ")
if len(password) >= 8 and ("@" in password or "#" in password):
    print("Strong password")
else:
    print("Weak Password")
```

```python
# 10 Pass / Fails
English = int(input("English: "))
Math = int(input("Math: "))
Science = int(input("Science: "))
if English>=40 and Math >= 40 and Science >= 40:
    print("Pass")
else:
    print("Fails")
```

```python
# 11 Positive and Even / Negative or Odd
num = int(input("Number: "))
if num>0 and num%2==0:
    print("Positive and Even")
elif num<0 or num % 2 != 0:
    print("Negative or Odd")
```

```python
# 12 City Name Operation
city = input("Enter City Name: ")
print("Reverse: ", city[::-1])
count = 0
for ch in city.lower():
    if ch in "aeious":
        count += 1
print("Vowels: ", count)
print("Without first and last: ",city[1:-1])
```
