```python
# 30 Interview Challenge
# Calculate each student's average marks.
students = [
    {"Name" : "Ram", "Marks" : [70, 80, 90]},
    {"Name" : "Amit", "Marks" : [90, 92, 94]},
    {"Name" : "Neha", "Marks" : [60, 65, 70]},
    {"Name" : "Raj", "Marks" : [88, 84, 90]}
]
average = list(map(lambda x : (x["Name"], sum(x["Marks"])/len(x["Marks"])), students))
print(average)           

# 1. Sort students by average marks (highest first).
sorted_students = sorted(students, key = lambda x : sum(x["Marks"])/len(x["Marks"]), reverse = True)
print(sorted_students)

# 2. Find the topper.
topper = max(students, key = lambda x : sum(x["Marks"])/len(x["Marks"]))
print(topper)

# 3. Filter students whose average is greater than 80.
filtered = list(filter(lambda x : sum(x["Marks"])/len(x["Marks"])>80, students))
print(filtered)

# 4. Print only the names of the filtered students.
names = list(map(lambda x : x["Name"], filtered))
print(names)
```

#### 4. Customer Order Processing

```python
orders = [
    {"id":1,"customer":"Ram","amount":4500,"status":"Delivered"},
    {"id":2,"customer":"Amit","amount":7000,"status":"Pending"},
    {"id":3,"customer":"Raj","amount":2000,"status":"Delivered"},
    {"id":4,"customer":"Neha","amount":9000,"status":"Delivered"},
    {"id":5,"customer":"Anu","amount":3500,"status":"Cancelled"},
    {"id":6,"customer":"Ravi","amount":10000,"status":"Delivered"},
]

# 1. Total delivered amount
total = sum(map(lambda x: x["amount"], filter(lambda x: x["status"]=="Delivered", orders)))
print(total)
```

```python
# 2. Average delivered amount
delivered = list(filter(lambda x : x["status"] == "Delivered",orders))
average = sum(map(lambda x: x["amount"], delivered))/ len(delivered)
print(average)
```

```python
# 3. Largest delivered order
largest = max(filter(lambda x:x["status"]=="Delivered", orders), key=lambda x:x["amount"])
print(largest)
```

```python
# 4. Pending orders
pending = list(filter(lambda x:x["status"]=="Pending",orders))
print(pending)
```

```python
# 5. Sort by amount descending
result = sorted(orders, key=lambda x:x["amount"], reverse = True)
print(result)
```

```python
# 6. Print only delivered customers
customer = list(map(lambda x:x["customer"], filter(lambda x: x["status"] == "Delivered", orders)))
print(customer)
```

```python
# 7. Count delivered orders
count = len(list(filter(lambda x: x["status"] == "Delivered", orders)))
print(count)
```

```python
# 8. Find customer spending maximum
maximum = max(orders, key = lambda x: x["amount"])
print(maximum)
```
